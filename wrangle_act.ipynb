### 定义通用函数

```python
def basicInfo(df):
    """
    The basicInfo function shows the describ info and the basic info.
    """
    des_tag="describ info"
    base_tag="basic info"
    shape_tag="shape info"
    row_count,col_count=df.shape
    #calculat the missed percent
    totalNumber = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()*100).round(decimals=2).sort_values(ascending=False)
    missing_data = pd.concat([totalNumber, percent], axis=1, keys=['MissedNumber', 'MissedPercent'])
    describ = df.describe().stack().unstack(0)
    dtype = DataFrame(df.dtypes,columns=['dtypes'])
    baseInfo = pd.concat([dtype,missing_data],axis=1)
    print(shape_tag.center(80,"*"))
    print(df.shape)
    print(des_tag.center(80,"*"))
    print(describ)
    #calculat the duplicated number
    #astype(str) for some colums is unhashable type
    duplicatedInfo =DataFrame(Series({x:df[x].astype(str).duplicated().sum() for x in df.columns}),columns=['DuplicatedNum'])
    baseInfo = pd.concat([baseInfo,duplicatedInfo],axis=1)
    print(base_tag.center(80,"*"))
    print(baseInfo)
```

#### 提取source数据

***定义:***

通过正则表达式提取source中的内容

***代码：***

```python
def getSource(x):
    '''
    提取sourcehheml中的数据，eg：提取“Twitter for iPhone”
    <a href="http://twitter.com/download/iphone" rel="nofollow">Twitter for iPhone</a>
    '''
    regex = r">(?P<content>.+)<"
    match = re.search(regex, x)
    return match.group("content")
```

```python
archive_df_copy["source"]=archive_df_copy.source.apply(getSource)
```

```python

def analyzeNumber(x):
    '''
    通过正则表达式从x中提取格式为"[0-9]+/[0-9]+""的数字
    '''
    regex = r"([0-9]+/[0-9]+)"
    matchs = re.finditer(regex, x)
    number = "/"
    for match in matchs:
        number = match.group(0)
        if(number.split("/")[1] =="10"):
            break
    return number.split("/")[0],number.split("/")[1]
```

#### 合并doggo，floofer，pupper，puppo

***定义：***
- 首先从text中通过正则表达进行提取stage信息
- 手动对于包含两个stage的信息进行修正（由于同时包含两个stage的数据较少）
- 最后删除这四类的数据

***代码：***

```python
archive_df_copy['stage']=""
```

```python
def findStage(x):
    '''
    从x中通过正则表达式提取符合要求的stage信息
    '''
    regex = r"((doggo[es|s]?)|(floofer[s]?)|(pupper[s]?)|(puppo[es|s]?))"
    matchs = re.finditer(regex, x,re.IGNORECASE)
    stage = ""
    stages=[]
    for match in matchs:
        stage = match.group(0)
        stages.append(stage.lower())
    return "_".join(set(stages)).replace(" ","")
```

```python
def updateRatingNumber(df,updateStageData):
    '''
    使用updateStageData中的数据更新df中对应的记录
    '''
    convertStruct={'puppos':"puppo",'puppers':"pupper",'doggos':'doggo',
                  'floofer_doggo':"doggo_floofer",'doggo_puppo':"puppo_doggo",'pupper_doggo':"doggo_pupper"}
    for index in updateStageData.index:
        df.loc[index,"stage"] = convertStruct.get(updateStageData[index].strip(),updateStageData[index])
```

```python
pp = archive_df_copy.text.apply(findStage)
```

尝试对text中的”This is “，”His name is“，”Hers is“，”That is“等前置expression之后的名词进行提取，通过如下函数中前置表达式进行提取。

```python
def findName(x):
    '''
    从x中通过正则表达式提取符合要求的stage信息
    '''
    regex = r"(?<=named)\s([A-Z][a-z]+)|(?<=His name is)\s([A-Z][a-z]+)|(?<=Here is)\s([A-Z][a-z]+)\
    |(?<=This is)\s([A-Z][a-z]+)|(?<=Her name is)\s([A-Z][a-z]+)|(?<=That is)\s([A-Z][a-z]+)"
    #regex = r"(?<=named\s)(\w)+"
    matchs = re.finditer(regex, x)
    name = ""
    for match in matchs:
        name = match.group(0)
        return name
```
